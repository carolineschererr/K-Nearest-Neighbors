```python
from sklearn import datasets
import pandas as pd
```

## Carregar Iris Dataset

```python
iris = datasets.load_iris()
```

## Criar Dataframe

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

## Gerar Alvo

```python
df['target'] = iris.target
```

```python
df
```

```python
iris.target_names
```

```python
df['target_names'] = iris.target_names[df['target']]
```

```python
df
```

## Determinar features

```python
iris_features = ['sepal length (cm)', 'sepal width (cm)',	'petal length (cm)',	'petal width (cm)']
```

## Determinar alvo

Quando se usa KNN, alvo deve ser numérico

```python
y = df.target
```

### Selecionar as features que serão utilizadas

```python
X = df[iris_features]
```

```python
X
```

```python
y
```

## Importar KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

### Selecionar modelo

```python
modelo = KNeighborsClassifier(3)
```

### Treinar modelo



```python
modelo.fit(X, y)
```

```python
X
```

### Ver predição

```python
print(modelo.predict(X))
```

### Checando acurácia

Com os mesmos dados em que o modelo foi treinado.

```python
modelo.score(X, y)
```
